```python
# Importing all the neccesary libraries
from captcha.image import ImageCaptcha
import random
import string
import os
from pathlib import Path
```

```python
# Generating a 1000 CAPTCHA Images inside Google Colab
data_path = Path("data/")
data_path.mkdir(exist_ok = True)
images_path = data_path / "images"
images_path.mkdir(exist_ok = True)

chars = string.ascii_lowercase + string.ascii_uppercase + string.digits

for i in range(1000):
  char = random.choice(chars)
  img = ImageCaptcha(width = 100,
                     height = 40)
  image = img.generate_image(char)
  image.save(images_path / f"{char}_{i}.png")
print(f"{len(os.listdir(os.path.join(images_path)))} images created")
```
